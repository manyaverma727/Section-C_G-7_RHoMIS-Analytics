## 5. Data Quality Issues Flagged for Cleaning

### Problem 1: Birth Year Contamination in Age Columns
- `age_malehead` and `age_femalehead` contain birth years (e.g., 1957, 1975) mixed with actual ages
- Mean = 109.9 for male head (inflated by birth years), median = 48.0
- Must convert using survey `year` column: `age = survey_year - birth_year`
- After conversion, any age >110 is irrecoverable garbage → set to NaN

### Problem 2: Country Name Inconsistencies
- `'mali'` (2,999 rows) and `'Mali'` (310 rows) as separate categories
- `'philipines'` is misspelled (should be `'philippines'`)
- Use `iso_country_code` as canonical reference; standardise `country` to lowercase

### Problem 3: unitland Contains 80+ Distinct Values
- Includes valid units: `'ha'`, `'hectare'`, `'acre'`, `'timad'` (0.25 ha), `'are'` (0.01 ha), `'m2'`
- Also contains garbage: dimension strings (`'50 by 100'`, `'100x100 feet'`), narratives, bare digits
- **Must build full conversion mapping before cleaning** — run `df['unitland'].value_counts()` in notebook 02

### Problem 4: landcultivated Extreme Values
- Max = 1,515,600 hectares (impossible) due to unit contamination
- Mean = 117 ha, median = 3.0 ha (mean is 39× median — skew = 214.08)
- Unit conversion and outlier capping required in notebook 02

### Problem 5: Currency Inconsistencies in local_currency
- CFA franc represented as: `'cfa_franc'`, `'fcfa_xof'`, `'fcfa'`, `'cfa'`, `'1'` (605 rows!)
- Missing in 20.3% of rows, making income cross-country comparison difficult
- Must standardise to canonical form (e.g., all CFA → `'xof'`)

### Problem 6: crop_land_area_* Columns Have Numeric Contamination
- Supposed to be categorical (`'little'`, `'underhalf'`, `'half'`, `'most'`, `'all'`, `'none'`)
- Contain numeric values (0–5,000) in several hundred rows
- Invalid entries must be set to NaN in cleaning

### Problem 7: offfarm_incomes_any Has Mixed Encoding
- Contains: `'y'`, `'n'`, `'TRUE'`, `'FALSE'`, `'no_answer'` as distinct values
- `'TRUE'`/`'FALSE'` encoding (498 + 438 rows) must be mapped to `'y'`/`'n'`

### Problem 8: HFIAS and FIES Are Structurally Missing, Not MCAR
- HFIAS: 69.7%–87.5% missing (instrument not deployed in all waves)
- FIES: 54.2% missing (most complete food security instrument — ~25,000 valid rows)
- Do not impute; retain as-is. They define separate sub-populations for analysis

## 6. Handoff to Notebook 02 — Structural Cleaning

**What notebook 02 will produce:**

A structurally clean, column-reduced dataset with:
1. **Column selection**: 60 working columns across 6 domains (identity, demographics, land, crops, income, food security)
2. **Unit standardisation**: `landcultivated` → `land_cultivated_ha` using full `unitland` conversion table
3. **Data entry error correction**: Birth-year ages converted using survey `year`; country/currency case standardised
4. **Categorical validation**: HFIAS/FIES strings cleaned; `crop_land_area_*` numeric contamination removed; binary fields (`offfarm_incomes_any`, `respondent_is_head`, etc.) standardised to `'y'`/`'n'`/NaN
5. **Household size derivation**: `household_size_derived` = sum of 8 age-band columns (0% missing, replaces imputed `count_people`)
6. **No analysis-phase outputs**: Food security classification, farm-size buckets, and per-capita KPIs are deferred to analysis notebooks

**What will NOT be in the cleaned output:**
- Analysis-phase derivations: `food_security_status`, `income_per_capita`, `farm_size_bucket`
- Columns with no defined role (e.g., instrument metadata)
- Currency-converted income (income remains in local currency; PPP conversion is analysis-phase)
